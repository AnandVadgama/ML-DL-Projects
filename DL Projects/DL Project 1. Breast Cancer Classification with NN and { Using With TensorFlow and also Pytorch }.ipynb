Splitting the data into training data & Testing data

Standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

**Building the Neural Network**

download.png

```python
# importing tensorflow and Keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers of Neural Network

model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),
                          keras.layers.Dense(20, activation='relu'),
                          keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# compiling the Neural Network

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the Meural Network

history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

Visualizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

Accuracy of the model on test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

# **Building Using Pytorch**


```python
import torch
```

```python
X_train_tensor = torch.from_numpy(X_train_std.astype(np.float32))
X_test_tensor = torch.from_numpy(X_test_std.astype(np.float32))
y_train_tensor = torch.from_numpy(Y_train.values.astype(np.float32))
y_test_tensor = torch.from_numpy(Y_test.values.astype(np.float32))
```

```python
X_test_tensor.shape
```

Define The Model

```python
from torch.utils.data import Dataset, DataLoader

class CustomDataset(Dataset):

  def __init__(self, features, labels):

    self.features = features
    self.labels = labels

  def __len__(self):

    return len(self.features)

  def __getitem__(self, idx):

    return self.features[idx], self.labels[idx]
```

```python
train_dataset = CustomDataset(X_train_tensor,y_train_tensor)
test_dataset = CustomDataset(X_test_tensor,y_test_tensor)
```

```python
train_dataloader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_dataloader = DataLoader(test_dataset, batch_size=32, shuffle=False)
```

```python
import torch.nn as nn

class Model(nn.Module):

  def __init__(self, num_features):
    super().__init__()
    self.linear = nn.Linear(num_features, 1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):
    out = self.linear(features)
    out = self.sigmoid(out)
    return out
```

```python
#important params

learning_rate = 0.1
epochs = 25
```

# Training pipeline

```python
# create model
model = Model(X_train_tensor.shape[1])

# defining optimizer
optimizer = torch.optim.SGD(model.parameters(), lr = learning_rate)

# loss calculate
loss_function = nn.BCELoss()

# *Training Pipeline*

# define loop
for epoch in range(epochs):
  for batch_features, batch_labels in train_dataloader:

    # forward pass
    y_pred = model(batch_features)

    # loss calculate
    loss = loss_function(y_pred, batch_labels.view(-1,1))

    # clears gradiets
    optimizer.zero_grad()

    # backward pass
    loss.backward()

    # params upgrade
    optimizer.step()

  # print loss in every epoch
  print(f'epoch: {epoch + 1}, loss: {loss}')
```

```python
# model evaluation with test_dataloader
model.eval()
accuracy_list= []
with torch.no_grad():
  for batch_features, batch_labels in test_dataloader:
    y_pred = model(batch_features)
    y_pred = (y_pred > 0.6).float()

    batch_accuracy = (y_pred.view(-1) == batch_labels).float().mean().item()
    accuracy_list.append(batch_accuracy)

overall = sum(accuracy_list) / len(accuracy_list)
print(f'Accuracy: {overall:.4f}')
```
