```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies_data = pd.read_csv('movies.csv')
```

```python
movies_data
```

```python
select_feature = ['genres','keywords','tagline','cast','director']
```

```python
for feature in select_feature:
    movies_data[feature] = movies_data[feature].fillna('')
```

```python
combined_feature = movies_data['genres'] + ' ' + movies_data['keywords'] + ' ' +  movies_data['tagline'] + ' ' + movies_data['cast'] + ' ' + movies_data['director']
combined_feature
```

```python
vectorizer = TfidfVectorizer()
```

```python
feature_vector = vectorizer.fit_transform(combined_feature)
```

```python
print(feature_vector)
```

```python
similarity = cosine_similarity(feature_vector)
```

```python
print(similarity)
```

```python
similarity.shape
```

```python
movie_name = input("enter your favourite movie name : ")
```

```python
list_of_all_title = movies_data['title'].tolist()
```

```python
list_of_all_title
```

```python
find_close_match = difflib.get_close_matches(movie_name, list_of_all_title)
```

```python
find_close_match
```

```python
close_match = find_close_match[0]
```

```python
close_match
```

```python
index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]
index_of_the_movie
```

```python
similarity_score = list(enumerate(similarity[index_of_the_movie]))
similarity_score
```

```python
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1],reverse = True)
```

```python
sorted_similar_movies
```

```python
print("Movies Suggested for You : \n")
i=1
for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = movies_data[movies_data.index==index]['title'].values[0]
    if (i<30):
        print(i,' ',title_from_index)
        i+=1
```

```python
movie_name = input("enter your favourite movie name : ")

list_of_all_title = movies_data['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_all_title)

close_match = find_close_match[0]

index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1],reverse = True)

print("Movies Suggested for You : \n")

i=1
for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = movies_data[movies_data.index==index]['title'].values[0]
    if (i<30):
        print(i,' ',title_from_index)
        i+=1
```
